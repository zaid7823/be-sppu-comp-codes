#### Implement **Gradient Descent** Algorithm to find the local minima of a function.

#### For example, find the local minima of the function $y = (x+3)^2$ starting from the point $x = 2$.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 6*x + 9   # (x + 3)^2

def df(x):
    return 2*x + 6          # 2(x + 3)
```

```python
# Gradient Descent Function
def gradientDescent(initial_x, learning_rate, iterations):
    x = initial_x
    x_history = [x] # keeps track of all x values of gradient descent

    for i in range(iterations):
        gradient = df(x)    # derivation = gradient
        x = x - learning_rate * gradient
        x_history.append(x)

    return x, x_history
```

```python
# Parameters for Gradient Descent
initial_x = 0
learning_rate = 0.1
iterations = 50

x_min, x_history = gradientDescent(initial_x, learning_rate, iterations)
print(f"Local Minima = {x_min:.2f}")
```

```python
# Plotting the function and Gradient Descent path

xValues = np.linspace(-4, 2, 100)

plt.plot(xValues, f(xValues), label="f(x) = (x + 3)^2") # Function Plot
plt.plot(x_history, f(np.array(x_history)), 'rx', label="Gradient Descent Path")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
